## **Data Visualization**<br>
#### Dataset from [Kaggle](https://www.kaggle.com/datasets/sakshigoyal7/credit-card-customers)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import drive
drive.mount('/content/drive')
```

### Load Data from Drive

```python
df = pd.read_csv('/content/drive/MyDrive/Data DS Bootcamp/BankChurners.csv')
# Check information
df.info()
df
```

### 1. Visualization for multiple columns

#### a) 'attrition_flag' column

```python
# Create donut graph for 'attrition_flag'
plt.figure(figsize=(10, 5))
attrition_counts = df['Attrition_Flag'].value_counts()
plt.pie(attrition_counts, labels=attrition_counts.index, autopct='%1.1f%%', startangle=90, colors=['red', 'green'], wedgeprops={'width':0.3})
plt.title('Attrition Flag Percentage')
plt.axis('equal')
plt.show()
```

There are 16.1% Customers flagged for attrition

#### b) 'customer_age' column

```python
# Create vertical bar chart for 'customer_age', colored by height
plt.figure(figsize=(10, 5))
plt.bar(df['Customer_Age'].value_counts().index, df['Customer_Age'].value_counts(), color='green')
plt.title('Customer Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

Most customers using the service is in the age range of 40 to 50

#### c) 'gender' column

```python
# Create graph for 'gender'
plt.figure(figsize=(10, 5))
plt.barh(df['Gender'].value_counts().index, df['Gender'].value_counts(), color=['pink', 'blue'])
plt.title('Gender Distribution')
plt.xlabel('Count')
plt.ylabel('Gender')
plt.grid(True)
plt.show()
```

Most customers that used the service are Female

#### d) 'education_level' column

```python
# Create a pie chart for 'educational_level' column
education_counts = df['Education_Level'].value_counts()
plt.figure(figsize=(10, 5))
plt.pie(education_counts, labels=education_counts.index, startangle=90, autopct='%1.1f%%')
plt.title('Education Level Percentage')
plt.axis('equal')
plt.show()
```

The most education level that most of the customers have the education level of Graduate (30.9%)<br>

A small proportion (4.5%) of the customers have the education level of Doctorate

#### e) 'income_category' column

```python
# Create a pie chart for 'income_category' column
income_counts = df['Income_Category'].value_counts()
plt.figure(figsize=(10, 5))
plt.pie(income_counts, labels=income_counts.index, startangle=90, autopct='%1.1f%%')
plt.title('Income Category Percentage')
plt.axis('equal')
plt.show()
```

Most customers (35.2%) is in the income category of 'less than $40K'<br>

A small proportion of customers (7.2%) is in the income category of '$120K +'

### 2. Countplot visualization

```python
# Countplot using sns
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='Gender', hue='Attrition_Flag')
plt.title('Attrition Flag Distribution')
plt.xlabel('Attrition Flag')
plt.ylabel('Count')
```

The proportion of "Existing Customer" and "Attrited Customer" for both gender is approximately the same

### 3. Scatterplot visualization

```python
# Scatterplot using sns
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, y='Months_on_book', x='Total_Trans_Amt', hue='Attrition_Flag', style='Attrition_Flag', palette='deep')
plt.title('Months on Book vs Total Transaction Amount')
plt.xlabel('Total Transaction Amount')
plt.ylabel('Months on Book')
```

All attrited customers raises from the customers who have the total transaction amount of under $11250, *regardless of the period of relationship with the bank*

### 4. Heatmap visualization

```python
# Heatmap visualization using sns
# Create dataframe for correlation
df_corr = df[['Customer_Age',
    'Months_on_book',
    'Credit_Limit',
    'Total_Revolving_Bal',
    'Total_Trans_Amt',
    'Total_Trans_Ct',
    'Avg_Utilization_Ratio',
    'Total_Amt_Chng_Q4_Q1',
    'Total_Ct_Chng_Q4_Q1']]

# Visualize the correlation heatmap
plt.figure(figsize=(12, 6))
sns.heatmap(df_corr.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
```

#### **Insights**
**Strong Positive Correlation**<br>
1) Customer Age and Months on Book (0.79)<br>
*Older customers tend to have been with the bank for longer time*<br>
2) Total Transaction Amount and Total Transaction Count (0.81)<br>
*More transaction mean higher total transaction amount*<br>
3)Total Revolving Balance and Average Utilization Ratio (0.62)<br>
*Higher balance means higher credit utilization*<br>
<br>
**Negative Correlation**<br>
1) Credit Limit and Average Utilization Ratio (-0.48)<br>
*The higher the credit limit, the lower the utilization ratio (maybe the customers did not spend near the maximum)*<br>
<br>
**Weak/No Correlation**<br>
1) Customer Age and Credit Limit<br>
*Customer age does not have any influence to the credit limit*
