```python
# 1. Buatlah program untuk mencetak tabel perkalian dari bilangan yang diberikan

number = int(input("Masukkan angka: "))
rentang = int(input("Masukkan rentang angka: "))
print(f"Tabel perkalian dari {number} dengan rentang {rentang}")
for n in range(1, rentang+1):
  print(f"{number} x {n} = {number*n}")
```

```python
# 2. Buatlah program untuk menghitung jumlah semua angka dari 1 ke angka yang diberikan

number = int(input("Masukkan angka: "))
sum = 0
for n in range(1, number+1):
  sum += n

print(f"Jumlah semua angka dari 1 ke {number} adalah {sum}")
```

```python
# 3. Buat program untuk menampilkan semua bilangan prima dalam suatu rentang

number = int(input("Masukkan angka: "))
rentang = int(input("Masukkan rentang angka: "))
list_angka = list(range(number, rentang+1))

bilangan_prima = []
for n in list_angka:
  if n>1: # bilangan prima terkecil = 2
    prima = True
    for i in range(2, n): # cek pembagian, bilangan prima = dapat dibagi 1 dan dirinya
      if n%i == 0:
        prima = False
        break
    if prima:
      bilangan_prima.append(n)

print("Bilangan prima:", bilangan_prima)
```

```python
# 6. Tulis fungsi Python untuk menemukan angka terbesar dari tiga angka.

def maksimum(a,b,c):
  """
  Mencari jumlah maksimum dari tiga angka
  Menggunakan fungsi max()
  """
  return max(a,b,c)

print("Angka terbesar:", maksimum(1,5,9))
```

```python
# 7. Tulis fungsi Python untuk menjumlahkan semua angka dalam list.

def total(numbers):
  """
  Menjumlahkan semua angka dalam list
  Menggunakan fungsi sum()
  """
  return sum(numbers)

baris_angka = [1,2,3,4,5]
print("Total dari baris angka:", total(baris_angka))
```

```python
# 8. Tulis fungsi Python untuk mengalikan semua angka dalam list.

def multiplikasi(numbers):
  """
  Mengalikan semua angka dalam list
  """
  hasil = 1
  for n in numbers:
    hasil = hasil*n
  return hasil

sample_list = [8,2,3,-1,7]
print("Hasil perkalian dari baris angka:", multiplikasi(sample_list))
```

```python
# 10.Tulis fungsi Python yang menerima string dan menghitung jumlah huruf besar dan huruf kecil.

def hitung_huruf_kapital_kecil(teks):
  """
  Menghitung jumlah huruf kapital dan kecil
  """
  huruf_kapital = 0
  huruf_kecil = 0

  for a in teks: # pengecekan setiap alfabet
    if a.isupper():
      huruf_kapital += 1 # penambahan ke huruf_kapital jika kapital
    elif a.islower():
      huruf_kecil += 1 # penambahan ke huruf_kecil jika kecil
  return huruf_kapital, huruf_kecil # mengembalikan 2 nilai (int)

teks = str(input('Masukkan teks: '))
huruf_kapital, huruf_kecil = hitung_huruf_kapital_kecil(teks)
print("Huruf kapital:", huruf_kapital)
print("Huruf kecil:", huruf_kecil)
```

```python
# 11. Tulis fungsi Python yang mengambil daftar dan mengembalikan list baru dengan elemen unik dari list pertama.

def nilai_unik(list):
  """
  Mengembalikan list baru dengan elemen unik dari list pertama
  """
  unique_list = []
  for a in list:
    if a not in unique_list:
      unique_list.append(a)
  return unique_list

list_sampel = [1,2,3,3,3,3,4,5]
print("List sampel:", list_sampel)
print("List unik:", nilai_unik(list_sampel))
```

```python
# 12. Tulis fungsi Python yang menggunakan angka sebagai input parameter dan periksa angkanya prima atau tidak

def is_prime(number):
  """
  Mengecek apakah bilangan adalah bilangan prima
  """
  if number>1:
    for i in range(2, number):
      if (number%i) == 0:
        return False
        break
    else:
      return True
  else:
    return False

is_prime(11)
```

```python
# 13. Tulis program Python untuk mencetak bilangan genap dari list yang diberikan.

# inisiasi list
sample_list = [1,2,3,4,5,6,7,8,9]
bil_genap = []
bil_ganjil = []
for n in sample_list: # pengecekan setiap nilai di dalam sample_list
  if n%2 == 0: # genap = dapat dibagi 2
    bil_genap.append(n)
  else:
    bil_ganjil.append(n)

print("Bilangan genap:", bil_genap)
print("Bilangan ganjil:", bil_ganjil)
```
