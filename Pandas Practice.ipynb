### **1. Menyisipkan kolom 'Full Name' setelah 'Last Name'**
#### Menggunakan .str.title() untuk mengkapitalisasi huruf pertama

```python
paid_transaction.insert(4, 'Full Name', (paid_transaction['First Name'].str.title() + ' ' + paid_transaction['Last Name'].str.title()))
paid_transaction.head(10)
```

### **2. Menyisipkan kolom 'GMV' setelah 'Seller Discount'**
#### Gross Merchandise Value (GMV) = transaction amount - seller discount + delivery fee

```python
paid_transaction.insert(9, 'GMV', (paid_transaction['Transaction Amount'] - paid_transaction['Seller Discount'] + paid_transaction['Delivery Fee']))
paid_transaction.head(10)
```

### **3. Pengelompokan 'Meta Category' menjadi 2 group dan membuat dataframe baru untuk menampilkan jumlah total GMV *bulanan* tiap group**
#### Group 1: Home/Food, Sport/Hobi/Otomotif, Fashion, Beauty/Health
#### Group 2: Gadget/ Komputer, Elektronik, Service/ Mokado, Babies/ Kids

```python
# Define group
group_1 = ['Home/Food', 'Sport/Hobi/Otomotif', 'Fashion', 'Beauty/Health']
group_2 = ['Gadget/ Komputer', 'Elektronik', 'Service/ Mokado', 'Babies/ Kids']

# Mapping menggunakan dictionary
category_map = {cat: 'Group 1' for cat in group_1}
category_map.update({cat: 'Group 2' for cat in group_2})

# Kolom baru 'Category' berisi pengelompokan berdasarkan nilai 'Meta Category'
paid_transaction['Category'] = paid_transaction['Meta Category'].map(category_map)
paid_transaction.head(10)
```

```python
# Mengubah format 'Paid Date' dari 'yyyymmdd' menjadi 'yyyy/mm/dd'
paid_transaction['Paid Date'] = pd.to_datetime(paid_transaction['Paid Date'].astype(str), format='%Y%m%d')

# Menambahkan kolom 'Month'
paid_transaction['Month'] = paid_transaction['Paid Date'].dt.to_period('M')
```

```python
# Membuat dataframe pivot table
# Pengelompokan berdasar 'Category'
# Menjumlahkan GMV per bulan dari 'Paid Date'
category_pivot = paid_transaction.pivot_table(values='GMV', index='Month', columns='Category', aggfunc='sum').reset_index()
category_pivot
```

### **4. Python statement untuk mencari Seller yang memiliki total GMV tertinggi pada 2017/08**
#### Menggunakan fungsi sum()

```python
# Menggabungkan dataframe "paid_transaction" dan "seller" berdasarkan "order number"
paid_seller_merged = pd.merge(paid_transaction, seller, on='Order Number')
paid_seller_merged.head(10)
```

```python
# Mencari 'Seller' dengan 'GMV' tertinggi pada 2017/08
highest_gmv = paid_seller_merged.loc[paid_seller_merged['Month'] == '2017/08'].groupby('Seller')['GMV'].sum().idxmax()
print("Seller dengan GMV tertinggi pada Agustus 2017: " + highest_gmv)
```

### **5. Python statement untuk mencari Seller yang memiliki transaksi terbanyak pada 2017/09**
#### Menggunakan fungsi count()

```python
# Monthly transaction amount for each seller
monthly_transaction_amount = paid_seller_merged.loc[paid_seller_merged['Month'] == '2017/09'].groupby('Seller')['Transaction Amount'].count()
#monthly_transaction_amount = paid_seller_merged.loc[paid_seller_merged['Seller']=="E-Bold"].groupby('Month')['Transaction Amount'].count()
monthly_transaction_amount
```

```python
# Mencari 'Seller' dengan total transaksi terbanyak pada 2017/09
highest_volume = paid_seller_merged.loc[paid_seller_merged['Month'] == '2017/09'].groupby('Seller').count().idxmax()
print("Seller dengan transaksi terbanyak pada September 2017: " + highest_volume)
```

### *Penjelasan fungsi*
#### 1) `.loc[ ]` = Melakukan filter dengan boolean value (True/False) terhadap tahun/bulan yang dipilih
#### 2) `.groupby().sum()` atau `.groupby().count()` = Mengelompokkan dan melakukan operasi matematika
#### 3) `.idxmax()` = Mencari yang tertinggi
